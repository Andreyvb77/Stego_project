**Генерация датасета**

```python
import os
from PIL import Image

# Создаем папку для сохранения изображений в Google Drive
output_dir = "/content/drive/MyDrive/FQW/grayscale_dataset"
os.makedirs(output_dir, exist_ok=True)

# Генерируем 256 изображений
for i in range(256):
    # Создаем изображение 256x256 пикселей
    img = Image.new('L', (256, 256), color=i)

    # Сохраняем изображение с именем вида gray_000.png, ... gray_ 255.png
    img.save(os.path.join(output_dir, f"gray_{i:03d}.png"))

    print(f"Создано изображение gray_{i:03d}.png с яркостью {i}")

print(f"\nГотово! Все 256 изображений сохранены в папке '{output_dir}'")
```
